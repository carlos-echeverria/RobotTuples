```python
from itertools import combinations
from collections import Counter, defaultdict
from random import choice, choices, seed
```

```python
def possibleSets(robots,k):
    # return list of all subsets of length k form the n-element array robots.
    return [frozenset(x) for x in combinations(robots,k)]


def addElementToGroup(totalSets,Group,discarded):
    # add element if constraints are met
    try:
        rElement = choice(totalSets)
        #rElement = totalSets[0]
        print("We choose a random element from the possible 3-tuples: {}.".format(rElement))

        Group.append(rElement)
        GroupL = [x for tup in Group for x in tup]
        cnt = Counter(GroupL)
        robotsInGroup = list(cnt.keys())
        if (cnt[1]<=4 and cnt[2]<=4 and cnt[3]<=4 and cnt[4]<=4 and cnt[5]<=4 and cnt[6]<=4 and cnt[7]<=4 and cnt[8]<=4 and cnt[9]<=4 and cnt[10]<=4 and cnt[11]<=4 and cnt[12]<=4 and cnt[13]<=4 and cnt[14]<=4 and cnt[15]<=4 and cnt[16]<=4 and cnt[17]<=4 and cnt[18]<=4):

            try:
                idx = totalSets.index(rElement)
                totalSets.pop(idx)
            except IndexError:
                print("error")
        else:
            discarded.append(Group.pop())
            idx = totalSets.index(rElement)
            totalSets.pop(idx)


    except IndexError:
        print("no more elements in first run")
        addElementToGroup(discarded, Group, discarded=[])
```

```python
n = 18 # number of robots
k = 3  # tuple size
robots = list(range(1,n+1))
groupSize = 24
seed(6)
```

```python
all_tuples = possibleSets(robots, k)
n_tuples = len(all_tuples)


print(f"\nWith {n} robots, there are {n_tuples} posible combinations of {k}-tuples.\n")
```

```python
iter = 0
Group1 = []
discarded=[]
totalSets = list(all_tuples)
while len(Group1)<groupSize and iter<n_tuples:
    iter=iter+1
    #print("Total Sets Remaining before iteration: {}".format(len(totalSets)))
    print("Iteration: {}".format(iter))
    print("Number of Tuples remaining before adding a new element: {}".format(len(totalSets)))
    print("The Group Before adding a new tuple is:\n {}.".format(Group1))
    addElementToGroup(totalSets, Group1, discarded)
    print("The Group After adding a new tuple is:\n {}.".format(Group1))
    Group1L = [x for tup in Group1 for x in tup]
    cntT = Counter(Group1L)
    print("Group Object {}.".format(cntT))
    print("Number of Tuples Remaining after iteration: {}.".format(len(totalSets)))
    #print("Sets Remaining:\n {}".format(totalSets))
    print("Number of Tuples Discarded so far: {}.".format(len(discarded)))
    #print("Sets Discarded:\n {}.".format(discarded))
    print("\n")
    if (len(Group1)==groupSize):
        print("GROUP WAS CREATED WITH THE DESIRED CONSTRAINTS.\n")
        totalSets=totalSets+discarded
    if (iter==n_tuples):
        print("\nERROR: A GROUP COULD NOT BE BUILT WITH THE GIVEN SETS.\n")
        break
```

```python
def count_elements_in_group(group):
    c = Counter()
    for t in group:
        c.update(t)
    return c


def is_complete_group(group, group_size, n_robots):
    c = count_elements_in_group(group)
    
    valid = (len(c) == group_size)
    for n in c.values():
        valid &= (n == n_robots)
    return valid
```

```python
def pair_key(pair):
    return tuple(sorted(list(pair[0]) + list(pair[1])))

def build_pairs_dict(available_tuples):
    tuples_pairs = defaultdict(list)
    for pair in combinations(available_tuples, 2):
        key = pair_key(pair)
        tuples_pairs[key].append(pair)
    return tuples_pairs
```

```python
def shuffle_pair_elements(current_group, all_tuples):
    # make a copy, I want to avoid modifying the input
    current_group = set(current_group)
    
    available_tuples = set(all_tuples) - set(current_group)
    available_tuples_pairs = build_pairs_dict(available_tuples)
    
    # pick two tuples at random
    replace_tuples = sample(current_group, k=2)

    # pick two tuples with the same elements to replace them
    available_matching_tuples = available_tuples_pairs[pair_key(replace_tuples)]
    if len(available_matching_tuples) == 0:
        # no tuples available, do nothing
        print('do nothing')
        return current_group

    new_tuples = choice(available_matching_tuples)
    print('replace this', replace_tuples)
    print('with this', new_tuples)
    
    # replace old tuples with new tuples
    current_group.remove(replace_tuples[0])
    current_group.remove(replace_tuples[1])
    current_group.add(new_tuples[0])
    current_group.add(new_tuples[1])
    
    return current_group
```
